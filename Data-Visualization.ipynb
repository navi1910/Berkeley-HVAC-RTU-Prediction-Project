## EDA - Visualization of the preprocessed Data

```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
sns.set_style('ticks')
from scipy.stats import ttest_rel
```

```python
# Loading the Pre-processed Data
df = pd.read_csv('datapreprocessed.csv', index_col = 'Timestamp')
```

```python
# Checking the shape of the Data
df.shape
```

```python
# Eliminating high-variation
df = df[(df['RTU: Circuit 1 Condenser Outlet Pressure']<600) & (df['RTU: Supply Air Volumetric Flow Rate'] > 3500) & (df['Terminal: Room 106 Air Humidity']>25)]
```

```python
# Checking the Top 5 observations of the Data
df.head()
```

```python
# Assigning variable to the Column names
columns = df.columns
columns
```

```python
# Heat Map for the Correlation matrix of the Pre-processed Data
plt.figure(figsize=(10,10))

ax = plt.axes()
sns.heatmap(df.corr())

ax.set_title('Heatmap - Preprocessed Data')
plt.show()
```

```python
# Boxplot describes the Target Variable
plt.figure(figsize = (6,4))
sns.boxplot(df['HVAC System: Electricity'], color = 'gold' ).set(title = 'Boxplot')
plt.show()
```

```python
# Remove Columns with Less Variation
df = df.drop(df[['Occupancy Mode Indicator','Lighting System: Control Command',
                  'RTU: Compressor 1 On/Off Status','VAV Box: Room 204 Reheat Status',
                  'VAV Box: Room 202 Reheat Status','VAV Box: Room 102 Reheat Status',
                  'VAV Box: Room 105 Reheat Status', 'VAV Box: Room 106 Reheat Status',
                  'VAV Box: Room 103 Reheat Status']],1)

# Plotting all the Discrete Variables
dis = list(df.select_dtypes(('int64')).columns)
print(dis)
for i in range(len(dis)):
    print(sns.countplot(df[dis[i]], palette='magma'))
    plt.title("0 - Off, 1 - On")
    plt.show()

# 0 - Off, 1 - On
```

#### Paired T-test 
To Check if there is statistically significant difference between Samples with Compressor 2 switched on/off

Null Hypothesis - No Significant difference betweeen the Samples
Alternate Hypothesis - There is Significant difference between the samples

```python
sns.violinplot(y = df['HVAC System: Electricity'], x = df['RTU: Compressor 2 On/Off Status'], palette = 'flag')
plt.title('0 - Off, 1 - On')
plt.show()

off_humi = df[df['RTU: Compressor 2 On/Off Status']==0]['HVAC System: Electricity'].sample(791)
on_humi = df[df['RTU: Compressor 2 On/Off Status']==1]['HVAC System: Electricity'].sample(791)

test_stat, p_value = ttest_rel(off_humi, on_humi)
print("P-value: ", p_value)
```

Since p_value less than 0.05 - Reject Null Hypthesis.

There is significant different between the Two Samples.

### There is a signifiant change in HVAC Systems Electricity consumption when Compressor 2 is turned Off/On.

### Checking for Correlation

```python
sns.jointplot(data = df, x = 'Lighting System: Electricity', y = df['HVAC System: Electricity'], kind = 'hex', color = 'green')
plt.show()
df[['Lighting System: Electricity', 'HVAC System: Electricity']].corr()
```

#### We can conclude that there is no correlation between Lighting System and HVAC System Electricity.

```python
sns.jointplot(data = df, x = 'RTU: Supply Air Temperature', y = df['HVAC System: Electricity'], color = 'blue')
plt.title('Jointplot')
plt.show()
df[['RTU: Supply Air Temperature','HVAC System: Electricity']].corr()
```

#### There is some amount of correlation between Supply Air Temperature and HVAC System Electricity Consumption.

```python
room_temps = ['Terminal: Room 102 Air Temperature',
       'Terminal: Room 103 Air Temperature',
       'Terminal: Room 104 Air Temperature',
       'Terminal: Room 105 Air Temperature',
       'Terminal: Room 106 Air Temperature',
       'Terminal: Room 202 Air Temperature',
       'Terminal: Room 203 Air Temperature',
       'Terminal: Room 204 Air Temperature',
       'Terminal: Room 205 Air Temperature',
       'Terminal: Room 206 Air Temperature']
```

```python
# Average Room Temperature is Calculated from All the Rooms
df['AverageRoomTemp'] = df[room_temps].sum(1).apply(lambda x:x/len(room_temps))
```

```python
print(df[['AverageRoomTemp','HVAC System: Electricity']].corr())
sns.jointplot(x = df['AverageRoomTemp'], y = df['HVAC System: Electricity'], color = 'red')
```

#### There is no correlation between Average Room Temperature and HVAC System Electricity.

```python
# Average Humidity in all available Rooms is calculated

humi = ['Terminal: Room 102 Air Humidity',
'Terminal: Room 104 Air Humidity',
'Terminal: Room 106 Air Humidity']

df['AverageRoomHumi'] = df[humi].sum(1)/len(humi)
```

```python
sns.jointplot(df['AverageRoomHumi'], df['AverageRoomTemp'], kind = 'kde', color = 'brown')
plt.show()
df[['AverageRoomHumi','AverageRoomTemp']].corr()
```

#### Slight correlation is seen between Average Room Humidity and Average Room Temperature.
